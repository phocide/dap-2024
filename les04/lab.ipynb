4. Выведите распределение переменной Pclass (социально-экономический статус) и это же распределение, только для мужчин / женщин по отдельности. Сколько было мужчин 2-го класса

6. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

```python
young = titanic_df[titanic_df['Age'] < 30]
young_survival_rate = young['Survived'].mean()

elderly = titanic_df[titanic_df['Age'] > 60]
elderly_survival_rate = elderly['Survived'].mean()

print(f"Выжившие моложе 30 лет: {young_survival_rate:.2%}")
print(f"Выжившие старше 60 лет:  {elderly_survival_rate:.2%}")

print(f"Люди моложе 30 лет выживали чаще, чем люди старше 60 лет?\nОтвет: {young_survival_rate > elderly_survival_rate}")
```

7. Правда ли, что женщины выживали чаще мужчин? Каковы доли выживших в обеих группах?

```python
women = titanic_df[titanic_df['Sex'] == 'female']
women_survival_rate = women['Survived'].mean()

men = titanic_df[titanic_df['Sex'] == 'male']
men_survival_rate = men['Survived'].mean()

print(f"Выжившие женщины: {women_survival_rate:.2%}")
print(f"Выжившие мужчины: {men_survival_rate:.2%}")

print(f"Женщины выживали чаще мужчин?\nОтвет: {women_survival_rate > men_survival_rate}")
```

8.  Найдите самое популярное имя среди пассажиров Титаника мужского пола?

```python
male = titanic_df[titanic_df['Sex'] == 'male'].copy()

male['FirstName'] = male['Name'].str.split('\\. ', expand=True)[1].str.split(' ', expand=True)[0]

male = male.dropna(subset=['FirstName'])

top_name = male['FirstName'].value_counts()
print(f"Самое популярное мужское имя: {top_name.index[0]} ({top_name.iloc[0]} раз)")
```

10. Сравните возраст у спасенных и у погибших пасажиров. Средний возраст погибших выше, верно?

```python
died = titanic_df[titanic_df['Survived'] == 0]
avg_age_died = died['Age'].mean()

survived = titanic_df[titanic_df['Survived'] == 1]
avg_age_survived = survived['Age'].mean()

print(f"Средний возраст погибших: {avg_age_died:.2f}")
print(f"Средний возраст выживших: {avg_age_survived:.2f}")
print(f"Средний возраст погибших выше? {avg_age_died > avg_age_survived}")
```
