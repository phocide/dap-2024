Ирисы Фишера — набор данных для задачи классификации, на примере которого Рональд Фишер в 1936 году продемонстрировал работу разработанного им метода дискриминантного анализа. Иногда его также называют ирисами Андерсона, так как данные были собраны американским ботаником Эдгаром Андерсоном.

```python
import sys
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

print(f"Python version: {sys.version}")
print(f"Numpy version: {np.version.version}")
print(f"Pandas version: {pd.__version__}")
print(f"Matplotlib version: {mpl.__version__}")
```

1. Импортируйте модель данных "Ирисы Фишера".

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

2. Выполните нормализацию данных

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

iris_scaled = scaler.fit_transform(iris.data)
```

3. Постройте модель кластеризации используя алгоритм  DBSCAN (https://www.youtube.com/watch?v=svAtnZ5XjSI&t=494s), подберите подходящие параметры eps, min_samples

```python
from sklearn.cluster import DBSCAN

model = DBSCAN(eps=0.8, min_samples=4)
y_pred = model.fit_predict(iris_scaled)
```

4. Постройте диаграмму рассеяния для признаков "sepal length (cm)" "sepal width (cm)", цвет точек на диаграмме должен быть различным для разных кластеров

```python
plt.figure(figsize=(10, 6))
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=y_pred, cmap='viridis')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('Диаграмма рассеяния кластеров (DBSCAN)')
plt.colorbar(label='Кластер')
plt.grid(True)
plt.show()
```

5. Сделайте вывод

Вывод:

Алгоритм DBSCAN успешно выявил три основных кластера в данных "Ирисы Фишера" по признакам длины и ширины. Два из них (зелёный и жёлтый) представляют собой плотные группы, соответствующие различным видам цветков, а третий кластер (фиолетовый) это точки, отнесённые к шуму. Диаграмма рассеяния показывает, что метод эффективно разделяет данные на компактные группы с четкими границами.
